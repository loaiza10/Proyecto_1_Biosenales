# **Introducción**

El presente proyecto se enmarca en el área de las Interfaces Cerebro–Computador (BCI, por sus siglas en inglés), cuyo objetivo principal es analizar señales electroencefalográficas (EEG) asociadas a tareas motoras reales e imaginadas, con el fin de evaluar la viabilidad de sistemas BCI no invasivos.

El estudio se basa en un conjunto de datos experimentales en el que múltiples sujetos realizan cuatro condiciones específicas: movimiento real de la mano derecha, movimiento real de la mano izquierda, imaginación del movimiento de la mano derecha e imaginación del movimiento de la mano izquierda, además de periodos de reposo. Estas señales EEG permiten explorar cómo la actividad cerebral varía según la intención o ejecución del movimiento.

El propósito central del proyecto es desarrollar un flujo de procesamiento de señales que permita identificar y caracterizar patrones en los ritmos cerebrales asociados a cada una de estas condiciones. A lo largo del proyecto se abordarán diferentes etapas, que incluyen:

- La exploración teórica de los fundamentos de las BCI y el análisis de señales EEG.
- El diseño de una metodología de procesamiento de señales que permita diferenciar las condiciones experimentales.
- La implementación de algoritmos para el cálculo de métricas relevantes, como la densidad espectral de potencia en distintas bandas de frecuencia.

La interpretación de los resultados mediante herramientas de estadística descriptiva e inferencial.

En conjunto, este proyecto busca integrar conocimientos de bioseñales, procesamiento digital y análisis de datos, con el objetivo de comprender mejor la dinámica cerebral asociada al control motor y su potencial aplicación en sistemas de Interfaz Cerebro-Computador (BCI) no invasivos. 

# **Historia, usos y proyectos futuros de las Interfaces Cerebro–Computador (BCI)**

Las Interfaces Cerebro–Computador (BCI) constituyen un campo interdisciplinario que integra neurociencia, ingeniería biomédica, procesamiento de señales y aprendizaje automático, con el objetivo de establecer una comunicación directa entre el cerebro y dispositivos externos.

**Historia de las BCI**

El desarrollo de las BCI tiene sus orígenes en la década de 1920, cuando Hans Berger registró por primera vez la actividad eléctrica cerebral mediante electroencefalografía (EEG). Sin embargo, no fue sino hasta la década de 1970 cuando se comenzaron a explorar aplicaciones orientadas a la comunicación cerebro–máquina.

Uno de los primeros avances significativos fue desarrollado por Jacques Vidal, quien introdujo el término Brain-Computer Interface y demostró la posibilidad de controlar dispositivos simples mediante señales EEG [1].

Durante las décadas de 1990 y 2000, el campo experimentó un crecimiento notable gracias al desarrollo de técnicas de procesamiento digital y algoritmos de clasificación. En este periodo se consolidaron paradigmas fundamentales como [2]:

* BCI basadas en potenciales evocados (P300)
* BCI basadas en ritmos sensoriomotores (SMR)
* BCI basadas en potenciales relacionados con eventos (ERP)

En años recientes, el avance de la inteligencia artificial y el aprendizaje profundo ha permitido mejorar significativamente la precisión y robustez de los sistemas BCI [3].

**Usos actuales de las BCI**

Las aplicaciones de las BCI se pueden clasificar en diferentes áreas:

**1. Rehabilitación neurológica**

Las BCI se utilizan para la recuperación funcional en pacientes con lesiones neurológicas, como accidente cerebrovascular o lesiones medulares. Mediante la imaginación motora, los pacientes pueden activar dispositivos de rehabilitación, favoreciendo la neuroplasticidad [4].

**2. Asistencia a personas con discapacidad**

Permiten a pacientes con parálisis severa (por ejemplo, síndrome de enclaustramiento) comunicarse o controlar dispositivos como [2]:

* Sillas de ruedas
* Prótesis robóticas
* Interfaces de escritura

**3. Control de dispositivos externos**

Las señales cerebrales pueden emplearse para controlar [3]:

* Brazos robóticos
* Drones
* Sistemas domóticos

**4. Neuroergonomía y videojuegos**

En contextos no clínicos, las BCI se utilizan para [3]:

* Evaluar estados cognitivos (atención, fatiga)
* Interacción en videojuegos
* Realidad virtual y aumentada

**5. Investigación neurocientífica**

Las BCI permiten estudiar la relación entre actividad cerebral y comportamiento, facilitando el análisis de procesos cognitivos y motores [3].

**Proyectos futuros y tendencias**

El futuro de las BCI está orientado hacia sistemas más precisos, accesibles y adaptativos. Algunas líneas de desarrollo incluyen:

* **BCI híbridas:** Integran múltiples modalidades (EEG, EMG, eye-tracking) para mejorar la robustez del sistema [5].

* **Integración con inteligencia artificial:** El uso de redes neuronales profundas permitirá mejorar la detección de patrones complejos en señales EEG.

* **Sistemas no invasivos de alta resolución:** Se busca mejorar la calidad de adquisición sin recurrir a técnicas invasivas.

* **Neuroprótesis avanzadas:** Desarrollo de prótesis controladas directamente por la actividad cerebral con retroalimentación sensorial.

**Empresas emergentes**

Proyectos como los de Neuralink y Kernel están impulsando interfaces de alta resolución con potencial clínico y comercial.

En conjunto, las BCI representan una tecnología clave para el futuro de la interacción humano–máquina, con aplicaciones que abarcan desde la medicina hasta el entretenimiento.

# **Actividad cerebral registrada por EEG en movimiento e imaginación motora**

La electroencefalografía (EEG) es una técnica que permite registrar la actividad eléctrica cerebral mediante electrodos ubicados en el cuero cabelludo (EEG estándar), constituyendo un método de exploración no invasivo. Adicionalmente, el registro puede realizarse a nivel de la superficie cortical (EEG cortical) o mediante electrodos intracerebrales (EEG de profundidad), dependiendo del nivel de invasividad requerido.

Las señales obtenidas reflejan la actividad de poblaciones neuronales corticales, especialmente de las neuronas piramidales, y se registran como diferencias de potencial entre electrodos. El EEG es una herramienta de gran utilidad clínica para el diagnóstico y seguimiento de diversas patologías, como la epilepsia, las encefalopatías y las alteraciones del estado de conciencia, entre otras [6].

Ademas, el EEG es ampliamente utilizado en sistemas de Interfaz Cerebro–Computador (BCI) debido a su alta resolución temporal, bajo costo y facilidad de implementación, lo que lo convierte en una herramienta adecuada para analizar procesos dinámicos como el control motor. [7]

Los sistemas BCI permiten interpretar las intenciones del usuario a partir de características extraídas de la actividad eléctrica cerebral, transformándolas en comandos capaces de controlar dispositivos externos. Este enfoque resulta especialmente relevante en pacientes con parálisis severa, al ofrecer una vía alternativa de comunicación e interacción con el entorno.

En particular, los sistemas BCI basados en EEG se centran en el estudio de la corteza sensoriomotora, debido a que esta región se activa tanto durante la ejecución de un movimiento como durante su imaginación. En este contexto, se ha demostrado que la ejecución real de un movimiento activa áreas motoras específicas, mientras que la imaginación del movimiento (motor imagery) genera patrones de activación similares, aunque sin producir respuesta muscular [7].

**Electrodos relevantes en EEG para BCI**

En estudios relacionados con el movimiento de las manos, los electrodos de mayor interés se localizan sobre la corteza sensoriomotora, de acuerdo con el sistema internacional 10–20 [7]. En particular, destacan:

* C3: asociado al hemisferio izquierdo (control de la mano derecha)
* C4: asociado al hemisferio derecho (control de la mano izquierda)
* Cz: ubicado en la línea media, correspondiente a la región central de la corteza motora

El cerebro presenta organización contralateral, por lo que cada hemisferio controla el lado opuesto del cuerpo. Esta característica permite identificar patrones de activación específicos en el EEG durante tareas motoras [7].

**Ritmos cerebrales de interés**

Los ritmos cerebrales de interés corresponden a patrones oscilatorios de la actividad eléctrica cerebral, registrados mediante EEG, que se caracterizan por presentarse en rangos específicos de frecuencia y estar asociados a funciones fisiológicas particulares. En el contexto de los sistemas BCI, estos ritmos son relevantes porque modulan su potencia en respuesta a tareas cognitivas o motoras, lo que permite utilizarlos como biomarcadores para identificar diferentes estados mentales [8].

En particular, en aplicaciones de imaginación motora, los ritmos de mayor interés son los ritmos sensoriomotores (SMR), que incluyen las bandas mu (μ) y beta (β). Estos ritmos presentan variaciones sistemáticas en su actividad —principalmente desincronización (ERD) y sincronización (ERS)— durante la ejecución o imaginación del movimiento, lo que permite diferenciar dichas condiciones a partir del análisis de señales EEG [7,9].

***Ritmo alfa (α)***
- Rango de frecuencia: 8–13 Hz
- Corresponde al ritmo de fondo predominante en el EEG de adultos sanos.
- Se observa principalmente en individuos despiertos en estado de relajación con los ojos cerrados.
- Presenta su máxima amplitud en regiones occipitales y disminuye con la apertura de los ojos.
- Su ausencia o asimetría entre hemisferios puede indicar alteraciones neurológicas [9]

***Ritmo Mu (μ)***
- Rango de frecuencia: 8–13 Hz
- Ubicación: región del surco central (corteza motora y somatosensorial primaria)
- Se caracteriza por aparecer en estado de reposo y bloquearse durante el movimiento o su imaginación, especialmente en la extremidad contralateral.
- Es particularmente prominente en áreas relacionadas con el control de la mano.
- Su supresión puede observarse incluso cuando el sujeto únicamente imagina la ejecución del movimiento [9].

***Ritmo Beta (β)***
- Rango de frecuencia: 14–30 Hz
- Ubicación: regiones frontocentrales
- Asociado a procesos de control motor, actividad cognitiva y atención.
- Durante la ejecución o imaginación del movimiento, presenta una disminución de potencia, seguida de un incremento posterior al movimiento.
- La asimetría interhemisférica en esta banda puede ser indicativa de alteraciones funcionales [9].

Las variaciones en el ritmo μ durante tareas de imaginación motora, en comparación con un estado de referencia, constituyen un indicador fiable de la activación cortical asociada a estas tareas. Por su parte, los cambios en la actividad β pueden estar relacionados con procesos cognitivos adicionales, como la toma de decisiones respecto al tipo de movimiento a imaginar y la frecuencia con la que este se ejecuta mentalmente [7].

**Fenómenos ERD y ERS**

El análisis de señales EEG en sistemas BCI se fundamenta en la evaluación de cambios dinámicos en la potencia de los ritmos cerebrales en respuesta a eventos específicos. En este contexto, los fenómenos de desincronización y sincronización reflejan modificaciones en el grado de organización temporal de la actividad neuronal [9].

- ***ERD (Event-Related Desynchronization):*** Se define como la disminución de la potencia espectral en una banda de frecuencia específica (principalmente μ o β), como resultado de una desincronización de la actividad neuronal. Este fenómeno indica un aumento en la activación cortical y se presenta durante la ejecución o imaginación del movimiento.
- ***ERS (Event-Related Synchronization):*** Corresponde al incremento de la potencia espectral en una banda determinada, asociado a una mayor sincronización neuronal. Generalmente ocurre después del movimiento y se interpreta como un proceso de recuperación o inhibición cortical.

Estos fenómenos constituyen indicadores fundamentales de la dinámica cerebral, ya que permiten identificar y diferenciar patrones de actividad asociados a tareas motoras. Por esta razón, el análisis de ERD y ERS representa la base para el desarrollo de sistemas BCI, al facilitar la detección de la intención de movimiento a partir de señales EEG [7,9].

**Coherencia en el Análisis de Señales EEG**

La **coherencia** es una métrica matemática fundamentada en el análisis de densidad espectral que permite cuantificar el nivel de sincronización o conectividad funcional entre dos o más regiones cerebrales. A diferencia de la correlación temporal, la coherencia se calcula en el **dominio de la frecuencia**, identificando zonas que oscilan en una banda de frecuencia común $f$ y cuya diferencia de fase $\Delta\phi$ se mantiene constante durante un intervalo de tiempo determinado. [aj],[bj]

Matemáticamente, para dos señales $x(t)$ y $y(t)$, la coherencia funcional $C_{xy}(f)$ se define como:

$$C_{xy}(f) = \frac{|P_{xy}(f)|^2}{P_{xx}(f)P_{yy}(f)}$$

Donde $P_{xy}(f)$ es la densidad espectral cruzada y $P_{xx}(f)$, $P_{yy}(f)$ son las densidades auto-espectrales. Un valor de $C_{xy} \approx 1$ indica una alta comunicación funcional, mientras que $C_{xy} \approx 0$ sugiere independencia entre las regiones.

**Limitaciones del Análisis de Coherencia**

Sin embargo, la aplicación de la coherencia tradicional en señales EEG presenta ciertas limitaciones:

1. **Conducción de Volumen (Efecto de propagación):** El cráneo, el líquido cefalorraquídeo y los tejidos son conductores eléctricos. Por tanto, una señal eléctrica fuerte generada en una región específica del cerebro se esparce y es captada por múltiples sensores simultáneamente. Esto genera que la coherencia registre una falsa sincronización (coherencia espuria) entre dos electrodos, cuando en realidad ambos están midiendo la misma fuente debido a la conducción eléctrica [cj].

2. **Falta de Direccionalidad:** Aunque la coherencia estadística permite confirmar que el área A y el área B están sincronizadas, es una medida simétrica; es decir, no indica en qué zona se inició la actividad, imposibilitando distinguir qué región emite la información y cuál la recibe [cj].

3. **Presunción de Estabilidad (No estacionalidad):** El cálculo matemático asume que las señales cerebrales se mantienen estables. Dado que la dinámica de las redes cerebrales cambia en fracciones de milisegundo, al utilizar ventanas de tiempo largas para el análisis, el sistema promedia la señal y puede omitir conexiones funcionales instantáneas o transitorias [cj].

4. **Influencia del Electrodo de Referencia:** Si el electrodo utilizado como referencia capta cualquier actividad eléctrica (ruido muscular o cardíaco), esta señal se inyectará en todos los canales activos. Al compartir este ruido común, la coherencia entre todos los electrodos aumentará artificialmente, distorsionando el mapa de conectividad real del sujeto.



# **Flujos de procesamiento en sistemas BCI y su implementación en el curso**

El desarrollo de sistemas BCI basados en EEG requiere la implementación de un flujo de procesamiento estructurado que permita transformar señales cerebrales crudas en información útil para la identificación de estados mentales. Este enfoque es coherente con los objetivos planteados en el proyecto del curso .

### **Flujo general de procesamiento en BCI**

Un sistema BCI típico se compone de varias etapas fundamentales:

* **Adquisición de la señal**

En esta etapa se registran señales EEG mediante múltiples electrodos ubicados en el cuero cabelludo, especialmente en regiones sensoriomotoras como C3, C4 y Cz. Estas señales se obtienen durante tareas de movimiento real, imaginación motora y reposo, y constituyen la base del sistema BCI [2].

* **Preprocesamiento**

Debido a la baja relación señal–ruido del EEG, es necesario aplicar técnicas de limpieza, tales como filtrado, eliminación de artefactos (parpadeo, actividad muscular, interferencias eléctricas) y técnicas como el análisis de componentes independientes (ICA) [2].

* **Segmentación (Epoching)**

La señal se divide en ventanas temporales o epochs, alineadas con eventos experimentales,como el inicio de la tarea, el periodo de imaginación o el reposo. Esto permite analizar la dinámica temporal de la actividad cerebral asociada a diferentes condiciones [3].

* **Extracción de características**

En esta etapa, las señales EEG se transforman en variables cuantificables que representan la actividad cerebral. Entre las más relevantes se encuentran:

  * Densidad espectral de potencia (PSD)
  * Potencia en bandas μ (8–13 Hz) y β (14–30 Hz)
  * Identificación de fenómenos ERD/ERS [3]
  * Métodos avanzados como transformadas (FFT, Wavelet) y matrices de covarianza [2], [7]

El objetivo es identificar patrones discriminativos entre diferentes tareas.

* **Clasificación y análisis**

Las características extraídas se utilizan para:

  * Comparar condiciones experimentales (mano izquierda vs derecha, movimiento vs imaginación, etc.)
  * Aplicar herramientas estadísticas o algoritmos de aprendizaje automático, como análisis discriminante lineal o máquinas de soporte vectorial [3], [7]

### **Flujo propuesto adaptado al proyecto del curso**

De acuerdo con los requerimientos del proyecto, el flujo de procesamiento que puede implementarse incluye:

* **Carga y exploración de datos:** lectura de señales EEG y selección de canales relevantes.

* **Preprocesamiento:** filtrado en bandas de interés (μ y β) y eliminación de ruido.

* **Segmentación:** división de la señal en epochs según condición experimental.

* **Extracción de características:** cálculo de la densidad espectral de potencia.

* **Construcción del dataframe:** organización de resultados por sujeto, condición y características.

* **Análisis estadístico:** comparación entre condiciones y validación de hipótesis.

### **Discusión**

El flujo propuesto es adecuado para el desarrollo del proyecto, ya que permite implementar un análisis completo de señales EEG utilizando herramientas computacionales accesibles. Además, este enfoque facilita la identificación de patrones asociados a la imaginación motora, lo cual constituye la base de los sistemas BCI [2].

Asimismo, la incorporación de métodos estadísticos y de aprendizaje automático permite evaluar la capacidad discriminativa de las características extraídas, contribuyendo al diseño de sistemas BCI más robustos y eficientes [3].

# **Metodología de procesamiento y análisis de señales EEG (Incluyendo Análisis de Coherencia)**

Con base en la información teórica recopilada sobre sistemas BCI y análisis de señales EEG, se propone la siguiente metodología para identificar diferencias en los ritmos cerebrales asociados a las cuatro condiciones de estudio: movimiento real e imaginación de movimiento de las manos derecha e izquierda. Este enfoque integra el análisis de ritmos sensoriomotores y el estudio de la conectividad funcional mediante coherencia [7], [9].

### **1. Selección de datos y canales de interés**
El análisis se enfoca en los electrodos **C3, C4 y Cz**, debido a su ubicación sobre la corteza sensoriomotora y su relevancia en el control motor contralateral [2], [7].

### **2. Preprocesamiento de la señal**

* **Filtrado pasa banda (8–30 Hz):** Para aislar los ritmos, $\mu$ (8-13 Hz) y $\beta$ (14-30 Hz), asociados a las tareas motoras [9].
* **Eliminación de artefactos:** Uso de técnicas de rechazo de segmentos o Análisis de Componentes Independientes (ICA) [2].
* **Normalización:** Para permitir comparaciones inter-sujeto.

### **3. Segmentación de la señal (Epoching)**

La señal se divide en ventanas temporales (*epochs*) etiquetadas según las cuatro condiciones experimentales y los periodos de reposo [3].

### **4. Extracción de características**

Para cada *epoch* y canal de interés (C3, C4, Cz), se calculan:

#### **4.1 Densidad espectral de potencia (PSD)**
Mediante el método de Welch para analizar la distribución de energía en frecuencia [2].

#### **4.2 Potencia en bandas de interés e Identificación de ERD/ERS**

Extracción de potencia promedio en bandas $\mu$ y $\beta$ para identificar fenómenos de desincronización (ERD) o sincronización (ERS) relacionada con eventos [9].

#### **4.3 Análisis de Coherencia (Conectividad Funcional)**
Como requerimiento para el estudio de conectividad, se calcula la coherencia de magnitud al cuadrado entre pares de electrodos (**C3-C4**, **C3-Cz** y **C4-Cz**). La coherencia $C_{xy}(f)$ permite cuantificar la sincronización de fase estable entre regiones:

$$C_{xy}(f) = \frac{|P_{xy}(f)|^2}{P_{xx}(f)P_{yy}(f)}$$

Este parámetro identifica si las áreas motoras están "comunicadas" durante la tarea, más allá de su potencia individual.

### **5. Organización de datos**
Se construye un dataframe estructurado donde cada fila representa un registro procesado, incluyendo:
* Sujeto y Condición.
* Potencia en bandas $\mu$ y $\beta$ por canal (C3, C4, Cz).
* **Coherencia en bandas $\mu$ y $\beta$** para los pares C3-C4, C3-Cz y C4-Cz.

### **6. Análisis de datos**
#### **6.1 Comparaciones principales y por canal**

Evaluación de lateralidad (C3 vs C4) y diferencias entre movimiento real e imaginación [7].

#### **6.2 Análisis de Conectividad**

Evaluación de la coherencia interhemisférica (C3-C4) para determinar si la comunicación entre cortezas motoras varía significativamente entre el reposo, el movimiento real y la imaginación motora.

### **7. Análisis estadístico**

Uso de estadística descriptiva (boxplots) y pruebas de hipótesis (t-test, Mann-Whitney, ANOVA) para determinar la significancia de los cambios en potencia y coherencia observados [3].

### **8. Interpretación de resultados**

* **Potencia:** Se espera observar ERD contralateral en las bandas $\mu$ y $\beta$ durante la actividad [9].
* **Coherencia:** Valores altos de coherencia ($C_{xy} \approx 1$) indicarán una fuerte sincronización funcional y comunicación entre las redes motoras involucradas en la tarea.

### **9. Resultado esperado**

Se espera evidenciar que la imaginación motora genera patrones de potencia y coherencia funcional similares al movimiento real, validando la viabilidad del sistema BCI y la utilidad de la coherencia como biomarcador de activación cerebral.

# **Hipótesis y herramientas de análisis**

### **Planteamiento de hipótesis**

Con base en la literatura sobre sistemas BCI y el comportamiento de los ritmos sensoriomotores durante tareas motoras e imaginación motora, se plantean las siguientes hipótesis:

### **Hipótesis general**

- **H0**: No existen diferencias significativas en la actividad de los ritmos sensoriomotores ($\mu$ y $\beta$) ni en la conectividad funcional entre las condiciones de movimiento real, imaginación motora y reposo.
- **H general:** Existen diferencias significativas en la actividad de los ritmos sensoriomotores ($\mu$ y $\beta$) entre las condiciones de movimiento real, imaginación de movimiento y reposo, las cuales pueden ser identificadas mediante el análisis de la potencia espectral y la conectividad funcional (coherencia) en las señales EEG [7], [8].

### **Hipótesis específicas**

- **ERD (real vs imaginación)**
  * **H0:** No existen diferencias significativas en la magnitud de la desincronización (ERD) entre el movimiento real y la imaginación motora.
  * **H1:** Existen diferencias en la magnitud de la desincronización (ERD) entre el movimiento real y la imaginación motora, siendo más pronunciada durante la ejecución real del movimiento [7], [8].

- **Discriminación por potencia espectral**
  - **H0:** La potencia espectral en las bandas $\mu$ y $\beta$ no permite discriminar entre las diferentes condiciones experimentales (mano izquierda, mano derecha, imaginación y reposo).
  - **H2:** La potencia espectral en las bandas $\mu$ y $\beta$ permite discriminar entre las diferentes condiciones experimentales (mano izquierda, mano derecha, imaginación y reposo), constituyéndose como una característica relevante para sistemas BCI [7].

- **Conectividad (coherencia C3–C4)**
  - **H0:** No existen diferencias significativas en la coherencia interhemisférica (C3–C4) en las bandas $\mu$ y $\beta$ entre las condiciones de reposo, movimiento e imaginación.
  - **H3:** Durante la ejecución e imaginación del movimiento, se altera la coherencia interhemisférica (C3-C4) en las bandas $\mu$ y $\beta$ respecto al estado de reposo, reflejando cambios en la comunicación funcional entre las cortezas motoras de ambos hemisferios.

### **Herramientas y metodología para la validación de hipótesis**

Para comprobar las hipótesis planteadas, se emplearán herramientas de procesamiento de señales y análisis estadístico, coherentes con el flujo de procesamiento descrito previamente.

Adicionalmente, se estimará la variación relativa de potencia respecto a un periodo de referencia (baseline) para cuantificar fenómenos de ERD/ERS [7], [9]. Ademas, se evaluará la sincronización entre pares de canales (C3-C4, C3-Cz, C4-Cz).

**Organización de los datos**

Los resultados se estructurarán en un DataFrame, que incluirá:
- Identificador del sujeto
- Condición experimental
- Canal / Par de canales
- Potencia en banda $\mu$ y $\beta$
- Coherencia en banda $\mu$ y $\beta$

Esta estructura facilitará el análisis estadístico posterior.

**Análisis estadístico**

Para evaluar las hipótesis, se emplearán herramientas de estadística descriptiva e inferencial:

* **Estadística descriptiva**
  * Media y desviación estándar
  * Diagramas de caja (boxplots)
  * Gráficas comparativas por condición

**Pruebas de hipótesis**

Dependiendo de la distribución de los datos y de la normalidad:
- **Pruebas paramétricas:** t-test para comparación entre dos grupos; ANOVA para múltiples condiciones.
- **Pruebas no paramétricas:** Mann–Whitney; Kruskal–Wallis.

Estas pruebas permitirán determinar si las diferencias observadas en potencia y coherencia son estadísticamente significativas.

### **Interpretación de resultados**

Los resultados se interpretarán en función de:
* La presencia de ERD en bandas $\mu$ y $\beta$ durante tareas motoras.
* La activación contralateral en electrodos C3 y C4.
* El nivel de sincronización (coherencia) entre regiones corticales.
* Las diferencias entre movimiento real e imaginación.

Esto permitirá validar la capacidad de las características extraídas para discriminar entre las condiciones, lo cual es fundamental en sistemas BCI [7], [8]. El planteamiento de estas hipótesis y la selección de herramientas aseguran un análisis riguroso de las señales EEG, cumpliendo con la evaluación de potencia y conectividad cerebral.

# **Análisis de resultados**

### **A. Estadística descriptiva**

* **Potencia μ en C3: Derecha vs Izquierda**



Las distribuciones de potencia en la banda μ para las condiciones de imaginación de la mano derecha e izquierda en el canal C3 presentan medianas muy similares (~2.2 × 10⁻¹¹), con amplia superposición de los rangos intercuartílicos y alta dispersión entre sujetos.

Esto indica que no existe una diferencia visual clara entre ambas condiciones. Aunque C3 está asociado al hemisferio izquierdo y debería reflejar mayor desincronización (ERD) durante el movimiento de la mano derecha (contralateral), la presencia de actividad ipsilateral de menor magnitud, junto con la variabilidad inter-sujeto, reduce la capacidad de discriminación cuando se analiza un único canal.

* **Potencia μ en C3: Actividad vs Reposo**



En esta comparación se observa una tendencia clara: la condición de reposo presenta mayores valores de potencia μ, mientras que durante la actividad (imaginación motora) la potencia disminuye.

Este comportamiento es consistente con el fenómeno de desincronización relacionada con eventos (ERD):

* En reposo → mayor sincronización neuronal → mayor potencia
* En actividad → desincronización → menor potencia

Aunque la diferencia no es extrema, sí es visualmente consistente entre sujetos, lo que sugiere un patrón fisiológico estable.

* **Comparación C3 vs C4 en banda μ**



C3 presenta una mediana y un rango intercuartílico notablemente más amplios que C4. La caja de C3 se extiende desde ~0.7 hasta ~3.5 × 10⁻¹¹, mientras que C4 es más compacta (~0.5 a ~2.7 × 10⁻¹¹). Esto indica que C3 captura más variabilidad en la banda μ, posiblemente reflejando mayor activación en el hemisferio izquierdo (motor de mano derecha), que podría estar influenciado con que es la mano dominante en la mayoría de los sujetos a nivel poblacional.

* **Comparación C3 vs C4 en banda β**



En la banda β ambos electrodos tienen medianas similares (~0.5 × 10⁻¹¹), pero C3 presenta outliers más extremos (hasta ~2.6 × 10⁻¹¹) que C4 (~2.2 × 10⁻¹¹). Las distribuciones son muy asimétricas (sesgo positivo), lo que ya anticipa la necesidad de pruebas no paramétricas. La diferencia entre C3 y C4 en β es menos pronunciada que en μ.

### **B. Pruebas estadísticas**

**Selección y justificación de las pruebas estadísticas**

Para las comparaciones entre dos condiciones se aplicaron de forma paralela el t-test pareado y la prueba de Wilcoxon pareada. El t-test pareado asume que las diferencias entre pares siguen una distribución normal, mientras que la prueba de Wilcoxon es no paramétrica y no impone este supuesto. La inspección visual de los boxplots reveló que las distribuciones de potencia en las bandas μ y β presentan asimetría positiva marcada y presencia de valores atípicos extremos en todos los canales y condiciones analizadas, lo que sugiere una violación del supuesto de normalidad. Por esta razón, la prueba de Wilcoxon pareada se considera la herramienta principal para la toma de decisiones estadísticas, y el t-test se utiliza como referencia complementaria. El hecho de que ambas pruebas coincidan sistemáticamente en dirección e interpretación en todos los casos refuerza la robustez de las conclusiones obtenidas. Para la evaluación simultánea de los efectos del tipo de tarea (movimiento real vs imaginación) y la condición de mano (derecha vs izquierda) se empleó un ANOVA de dos factores, que permite evaluar tanto los efectos principales como la interacción entre ambos factores sobre la potencia espectral registrada en C3.

**Banda μ (8–13 Hz)**

- **Comparación 1: Derecha vs Izquierda en C3**

La comparación entre el movimiento de la mano derecha y el movimiento de la mano izquierda en el electrodo C3 no arrojó diferencias estadísticamente significativas. La prueba de Wilcoxon pareada obtuvo un valor p = 0.426, y el t-test pareado un valor p = 0.524, ambos ampliamente superiores al umbral de α = 0.05. Por lo tanto, no se rechaza la hipótesis nula: la potencia en la banda μ registrada en C3 no permite discriminar entre el movimiento de la mano derecha y el de la mano izquierda cuando se analiza de forma aislada.

Este resultado es coherente con la teoría de organización contralateral del sistema nervioso motor. Aunque C3 se asocia al hemisferio izquierdo y debería mostrar ERD contralateral selectivo durante el movimiento de la mano derecha, el cálculo de potencia promedio sobre todos los ensayos y sujetos mezcla los efectos de ERD contralateral e ipsilateral, diluyendo la diferencia entre condiciones. La variabilidad inter-sujeto en la magnitud del ERD supera la diferencia sistemática entre las dos condiciones cuando se analiza un único canal de forma independiente. La discriminación correcta de la lateralidad del movimiento requiere comparar C3 y C4 de forma simultánea, aprovechando el patrón espacial diferencial que genera la organización contralateral del sistema motor.

- **Comparación 2: Actividad vs Reposo en C3**

La comparación entre los periodos de actividad motora y el estado de reposo en C3 arrojó diferencias estadísticamente significativas en la banda μ. La prueba de Wilcoxon pareada obtuvo p = 0.0009 y el t-test pareado p = 0.043, ambos por debajo de α = 0.05. Se rechaza la hipótesis nula: la potencia en la banda μ es significativamente diferente entre la actividad y el reposo, siendo menor durante la actividad.

Este resultado confirma la presencia del fenómeno ERD en la banda μ sobre la corteza sensoriomotora. Desde el punto de vista fisiológico, el ritmo μ es generado por la actividad sincrónica de poblaciones neuronales de la corteza sensoriomotora primaria durante el estado de reposo. Cuando el sujeto ejecuta o imagina un movimiento, estas neuronas se dessincronizan para procesar información motora, lo que se traduce en una caída de la potencia espectral en la banda de 8 a 13 Hz. Este bloqueo del ritmo μ es uno de los biomarcadores más utilizados en sistemas BCI porque ocurre tanto durante el movimiento real como durante la imaginación motora, lo que permite usar la imaginación como señal de control sin necesidad de movimiento físico. La significancia estadística obtenida, especialmente el valor p = 0.0009 de la prueba de Wilcoxon, indica que esta diferencia es robusta y consistente entre sujetos, validando que el flujo de procesamiento implementado es capaz de detectar de forma confiable la modulación de la actividad cerebral asociada a las tareas motoras.

- **Comparación 3: C3 vs C4 en banda μ**

La comparación entre los electrodos C3 y C4 en la banda μ arrojó las diferencias más significativas de todo el análisis. La prueba de Wilcoxon pareada obtuvo p = 7.96 × 10⁻⁶ y el t-test pareado p = 0.00047, ambos muy por debajo del umbral de significancia (α = 0.05). Por lo tanto, se rechaza la hipótesis nula con alta confianza: existen diferencias estadísticamente significativas en la potencia de la banda μ entre ambos hemisferios.

Desde el punto de vista fisiológico, este resultado refleja la organización contralateral del sistema motor. Durante la imaginación o ejecución del movimiento de la mano derecha, se produce desincronización (ERD) en C3 (hemisferio izquierdo), mientras que para la mano izquierda ocurre en C4 (hemisferio derecho). Sin embargo, al integrar múltiples sujetos y condiciones, estas dinámicas no se compensan perfectamente.

La diferencia observada, donde C3 presenta mayor potencia promedio que C4, puede explicarse por varios factores:

* Predominancia de sujetos diestros, lo que implica una mayor representación cortical del hemisferio izquierdo.
* Diferencias en la magnitud del ERD entre hemisferios (no necesariamente simétricas).
* Variabilidad inter-sujeto en la activación motora y en la imaginación.

En conjunto, esto genera una asimetría global en la potencia de la banda μ.


**Banda β (14–30 Hz)**

- **Comparación 1: Derecha vs Izquierda en C3**

Al igual que en la banda μ, la comparación entre el movimiento de la mano derecha y la mano izquierda en C3 no reveló diferencias significativas en la banda β. La prueba de Wilcoxon arrojó p = 0.626 y el t-test p = 0.480, ambos muy por encima de α = 0.05. No se rechaza la hipótesis nula.

La banda β presenta un ERD más difuso espacialmente que el ritmo μ, extendiéndose hacia regiones frontocentrales y siendo menos lateralizado. Esto hace que la diferencia entre condiciones de mano sea aún menos detectable en un único electrodo que en la banda μ. Además, la banda β está asociada a procesos cognitivos generales como la atención y la preparación del movimiento, que son compartidos entre las condiciones de mano derecha e izquierda, reduciendo la especificidad lateral de la señal. La conclusión es consistente con la de la banda μ: discriminar la lateralidad del movimiento requiere el análisis conjunto de C3 y C4.

- **Comparación 2: Actividad vs Reposo en C3**

La comparación entre actividad y reposo en la banda β resultó significativa. La prueba de Wilcoxon obtuvo p = 0.0001 y el t-test p = 0.028, ambos por debajo de α = 0.05. Se rechaza la hipótesis nula: la potencia en la banda β también disminuye de forma significativa durante la actividad motora respecto al reposo.
Este resultado confirma el fenómeno ERD en la banda β, que según la literatura ocurre durante la ejecución e imaginación del movimiento y se interpreta como un indicador de activación del circuito motor. La banda β es especialmente sensible a los procesos de planificación y control motor: su desincronización durante la tarea refleja el reclutamiento activo de redes frontocentrales involucradas en la selección y coordinación del movimiento. El valor p = 0.0001 obtenido con Wilcoxon es incluso más bajo que el de la banda μ para la misma comparación. 

- **Comparación 3: C3 vs C4 en banda β**

La comparación entre C3 y C4 en la banda β también arrojó diferencias significativas. La prueba de Wilcoxon obtuvo p = 0.0002 y el t-test p = 0.034. Se rechaza la hipótesis nula: existe asimetría interhemisférica en la banda β, con C3 presentando mayor potencia que C4, aunque la diferencia es de menor magnitud que en la banda μ.

Aunque la banda β es más difusa espacialmente, la asimetría hemisférica asociada a la dominancia manual también se manifiesta en esta banda. La mayor representación cortical de la mano derecha en el hemisferio izquierdo produce señales β de mayor potencia en C3, de forma análoga a lo observado en μ. Sin embargo, la diferencia es menos pronunciada porque la banda β integra contribuciones de regiones más amplias, incluyendo áreas prefrontales y premotoras, que son más simétricas entre hemisferios. La distribución marcadamente asimétrica con outliers extremos observada en los boxplots de β, y el hecho de que el valor p del t-test (0.034) sea más cercano al umbral que el de Wilcoxon (0.0002), respaldan la elección de la prueba no paramétrica como la más apropiada para este conjunto de datos.

**ANOVA de dos factores — banda μ y β en C3**



- **Banda μ en C3**

El ANOVA de dos factores evaluó simultáneamente el efecto del tipo de tarea (movimiento real vs imaginación motora) y de la condición de mano (derecha vs izquierda) sobre la potencia en la banda μ registrada en C3, así como la interacción entre ambos factores. Ninguno de los tres términos del modelo alcanzó significancia estadística al nivel de α = 0.05. El factor Tipo obtuvo F = 2.857 con p = 0.091, el factor Condición F = 1.007 con p = 0.316, y la interacción Tipo × Condición F = 0.123 con p = 0.726.

El resultado del factor Tipo merece una interpretación especial. Con F = 2.857 y p = 0.091, se encuentra en una zona de tendencia marginal que no alcanza el umbral convencional pero que tampoco puede descartarse como ruido. Desde la perspectiva fisiológica, la literatura reporta consistentemente que el ERD durante el movimiento real es de mayor magnitud que durante la imaginación motora, dado que el movimiento real involucra tanto la activación cortical como la ejecución del comando motor efectivo hacia la médula espinal. La tendencia observada sugiere que esta diferencia existe en los datos, pero que la potencia estadística del análisis de un único electrodo, combinada con la alta variabilidad inter-sujeto característica de los estudios de imaginación motora, es insuficiente para alcanzar significancia formal. Este resultado no invalida la hipótesis de que el movimiento real y la imaginación generan patrones distintos, sino que señala que detectar esta diferencia requiere mayor cantidad de sujetos, análisis multicanal o métodos de extracción de características más sensibles.

El factor Condición (p = 0.316) y la interacción (p = 0.726) confirman que C3 aislado no diferencia la mano usada ni muestra un patrón de interacción entre tipo de tarea y mano, lo que es consistente con los resultados de las pruebas pareadas previas.

- **Banda β en C3**

En la banda β, el ANOVA de dos factores no encontró efectos significativos en ninguno de los términos evaluados. El factor Tipo obtuvo F = 0.408 con p = 0.523, el factor Condición F = 1.471 con p = 0.225, y la interacción F = 0.040 con p = 0.842. Todos los valores p están claramente por encima de α = 0.05, y los valores F son bajos, lo que indica que ninguna de las fuentes de variación evaluadas explica de forma apreciable la varianza en la potencia β de C3.

Este resultado es coherente con la naturaleza más difusa y menos lateralizada de la banda β respecto a μ. Dado que β integra contribuciones de regiones corticales más amplias y está asociado a procesos cognitivos generales compartidos entre condiciones, es esperable que las cuatro condiciones experimentales generen patrones de potencia β en C3 que no se diferencian de forma significativa cuando se analiza el canal de forma aislada. La ausencia completa de tendencias marginales en β, a diferencia de la tendencia observada en μ para el factor Tipo, refuerza que la banda μ es más específica para capturar diferencias relacionadas con la naturaleza de la tarea motora en electrodos sensoriomotores.

# **Conclusiones**

* **Validación del fenómeno ERD en bandas sensoriomotoras**
La comparación entre actividad y reposo arrojó los resultados más contundentes del análisis. En la banda μ (Wilcoxon p = 0.0009) y en la banda β (Wilcoxon p = 0.0001) se confirmó que la potencia espectral disminuye significativamente durante la ejecución o imaginación del movimiento en comparación con el estado de reposo. Este hallazgo valida la presencia del fenómeno ERD en la corteza sensoriomotora y constituye el fundamento fisiológico más sólido del proyecto: el cerebro modula de forma detectable su actividad oscilatoria ante la intención o ejecución de un movimiento, lo que es precisamente el principio que hace posible un sistema BCI basado en EEG. El ritmo μ, que se bloquea selectivamente durante el movimiento y su imaginación, y el ritmo β, asociado al control motor activo y a la toma de decisiones sobre el movimiento, se comportan como biomarcadores confiables de activación cortical motora.

* **La lateralidad de la mano no es discriminable con un solo electrodo**
En ninguna de las bandas analizadas (μ ni β) se encontraron diferencias significativas entre el movimiento de la mano derecha y la mano izquierda cuando se analiza exclusivamente el electrodo C3. Esto no contradice la teoría de organización contralateral del sistema motor; por el contrario, la confirma: para discriminar qué mano se está moviendo o imaginando, es necesario comparar C3 con C4 de forma simultánea. El ERD contralateral en C3 para la mano derecha y en C4 para la mano izquierda, junto con la ausencia de ERD (o ERS) en el hemisferio ipsilateral, generan un patrón espacial diferencial que no puede ser capturado por la potencia univariada de un solo canal. Un clasificador BCI funcional debe explotar esta asimetría interhemisférica como característica discriminativa central.

* **Asimetría interhemisférica significativa entre C3 y C4**
Las diferencias entre C3 y C4 resultaron altamente significativas tanto en μ (p = 7.96 × 10⁻⁶) como en β (p = 0.0002), con C3 mostrando mayor potencia y variabilidad en ambas bandas. Esta asimetría refleja la mayor representación cortical de la mano derecha (dominante en la mayoría de los sujetos) en el hemisferio izquierdo, y confirma que los electrodos C3 y C4 están capturando actividad fisiológica diferenciada. Este resultado es el más importante para el diseño de un sistema BCI: demuestra que existe información discriminativa entre hemisferios que puede ser aprovechada para clasificar la intención de movimiento.

* **El análisis multifactorial revela los límites del enfoque univariado**
El ANOVA de dos factores confirmó que ningún factor aislado en C3 (tipo de tarea, condición de mano, ni su interacción) alcanza significancia estadística para discriminar las cuatro condiciones experimentales. La tendencia marginal observada para el tipo de tarea en μ (p = 0.091) sugiere que existe una diferencia real entre movimiento real e imaginación, pero que la potencia de un solo electrodo no es suficiente para detectarla de forma robusta ante la variabilidad inter-sujeto. Esto indica que el enfoque de análisis debe evolucionar hacia estrategias multicanal: pares de electrodos (C3–C4), métricas de conectividad como la coherencia interhemisférica, y eventualmente métodos de clasificación supervisada que integren múltiples características simultáneamente.

* **Implicaciones para la viabilidad del sistema BCI**
El conjunto de resultados valida parcialmente la hipótesis general del proyecto. El sistema BCI no invasivo basado en EEG es viable para distinguir estados de actividad versus reposo, con alta significancia estadística en ambas bandas sensoriomotoras. La distinción entre las cuatro condiciones (movimiento real e imaginado de cada mano) requiere análisis más sofisticados que la potencia espectral univariada. Los próximos pasos naturales para mejorar la discriminación incluyen: calcular índices de lateralización (potencia C3 – potencia C4), incorporar la coherencia interhemisférica en bandas μ y β como medida de conectividad funcional, y aplicar clasificadores que combinen múltiples características para discriminar de forma robusta entre las cuatro condiciones experimentales.
